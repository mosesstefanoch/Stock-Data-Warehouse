```python
listGudang = [
    #Nama barang, satuan, banyak barang, harga barang /satuan
    ['Beras Putih', 'kg', 50, 25000],
    ['Gula Pasir', 'kg', 7, 14500],
    ['Saos Sambal', 'kg', 18, 22000]
]
```

```python
def lihatIsiStok():
    print('Daftar Stok barang\n')
    print('Index\t| Nama\t\t| Satuan\t| Stock\t| Harga per Satuan')
    for i in range(len(listGudang)) :
        print('{}\t| {}\t| {}\t\t| {}\t| {}'.format(i,listGudang[i][0],listGudang[i][1],listGudang[i][2],listGudang[i][3]))
```

```python
def tambahIsiStok():
    while True:
        namaBarang = input('Masukkan Nama Barang : ')
        satuanBarang = input('Masukkan Satuan Barang : ')
        stokBarang = int(input('Masukkan Stok Barang : '))
        hargaBarang = int(input('Masukkan Harga Barang : '))
        pilihan = input('''
        Barang yang ingin ditambahkan:
        
        Nama\t\t| Satuan\t| Stock\t| Harga per Satuan
        {}\t| {}\t\t| {}\t| {}
        
        Apakah sudah benar ? (y/n):
        '''.format(namaBarang, satuanBarang, stokBarang, hargaBarang))
        
        if(pilihan == 'y'):
            listGudang.append([namaBarang, satuanBarang, stokBarang, hargaBarang])
            lihatIsiStok()
            break
        elif(pilihan == 'n'):
            break
```

```python
def updateStok():
    while True: 
        lihatIsiStok()
        try:
            indexBarang = int(input('Masukkan Index Barang yang Ingin Diupdate: '))
            
            if(len(listGudang)-1 < indexBarang):
                print('Index barang tidak ditemukan!')
                continue
            else:
                namaBarang = input('Masukkan Nama Barang Baru : ')
                satuanBarang = input('Masukkan Satuan Barang Baru : ')
                stokBarang = int(input('Masukkan Stok Barang Baru : '))
                hargaBarang = int(input('Masukkan Harga Barang Baru : '))
                
                pilihan = input('''
                Barang ini:

                Nama\t\t| Satuan\t| Stock\t| Harga per Satuan
                {}\t| {}\t\t| {}\t| {}
                
                Akan diganti menjadi:
                
                Nama\t\t| Satuan\t| Stock\t| Harga per Satuan
                {}\t| {}\t\t| {}\t| {}
                
                Apakah sudah benar ? (y/n):
                '''.format(listGudang[indexBarang][0],listGudang[indexBarang][1],listGudang[indexBarang][2],listGudang[indexBarang][3],
                          namaBarang, satuanBarang, stokBarang, hargaBarang))

                if(pilihan == 'y'):
                    listGudang[indexBarang][0] = namaBarang
                    listGudang[indexBarang][1] = satuanBarang
                    listGudang[indexBarang][2] = stokBarang
                    listGudang[indexBarang][3] = hargaBarang
                    lihatIsiStok()
                    break
                elif(pilihan == 'n'):
                    break
        except ValueError:
            print("Input Invalid!")
```

```python
def hapusStok():
     while True: 
        lihatIsiStok()
        try:
            indexBarang = int(input('Masukkan Index Barang yang Ingin Dihapus: '))
            
            if(len(listGudang)-1 < indexBarang):
                print('Index barang tidak ditemukan!')
                continue
            else:
                pilihan = input('''
                Barang yang ingin dihapus:

                Nama\t\t| Satuan\t| Stock\t| Harga per Satuan
                {}\t| {}\t\t| {}\t| {}

                Apakah sudah benar ? (y/n):
                '''.format(listGudang[indexBarang][0],listGudang[indexBarang][1],listGudang[indexBarang][2],listGudang[indexBarang][3]))

                if(pilihan == 'y'):
                    del listGudang[indexBarang]
                    lihatIsiStok()
                    break
                elif(pilihan == 'n'):
                    break
        except ValueError:
            print("Input Invalid!")
```

```python
while True:
    menu = input('''
    ============GUDANG SERBA ADA============
    
    Selamat datang silahkan pilih menu dibawah ini:
    1. Menambahkan stok barang
    2. Melihat isi stok barang
    3. Mengupdate stok barang
    4. Menghapus stok barang
    5. Exit
    
    Masukkan menu yang ingin diinginkan: 
    ''')
    
    if(menu == '1') :

        tambahIsiStok()

    elif(menu == '2') :

        lihatIsiStok()

    elif(menu == '3') :

        updateStok()

    elif(menu == '4') :

        hapusStok()
                
    elif(menu == '5') :
        print('''
        Terimakasih sampai jumpa kembali :D
        ''')
        break
```
